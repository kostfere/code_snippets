```python
import numpy as np
import pandas as pd
from scipy import optimize
```

```python
nr_mults = np.random.randint(low=3, high=5)
dims_per_mult = np.random.randint(low=2, high=3, size=nr_mults)
```

```python
muls=list()
dims=list()
for mul in range(0,nr_mults):
    for dim in range(0,dims_per_mult[mul]):
        muls.append('mul_'+str(mul))
        dims.append('dim_'+str(dim))
df = pd.DataFrame(data = {'muls': muls,
                         'dims': dims}
                  , columns=['muls','dims'])
#df = df[['muls','dims']]
df['code'] = ['w['+str(i)+']' for i in range(0,len(df))]

uniq_mults =np.unique(df.muls)

df_fin = df[df.muls==uniq_mults[0]][['code']].rename(columns={'code':'code_'+uniq_mults[0]})
df_fin.loc[:, 'dummy_var']=1;
for mul in uniq_mults[1:]:
    df_temp=df[df.muls==mul][['code']].rename(columns={'code':'code_'+mul})
    df_temp.loc[:, 'dummy_var']=1
    df_fin = df_fin.merge(df_temp,on = 'dummy_var')
df_fin = df_fin.drop('dummy_var',axis=1)
df_fin.loc[:, 'prod_format'] = df_fin.apply(lambda row: '*'.join(map(str, row.values)), axis=1)
k_values_str = np.random.uniform(0,2,size=len(df_fin))+2
df_fin.loc[:, 'k_values'] = k_values_str
```
